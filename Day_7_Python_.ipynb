**1. NumPy (Numerical Python)**

- NumPy provides support for **multi-dimensional arrays** and **mathematical operations** on them.  
- Regular Python lists are slow and not suitable for numerical operations. NumPy arrays are fast, memory efficient, and powerful.  
- Used in Machine Learning, AI, Image Processing, and Data Science.

**Example:**  


```python
import numpy as np

# Create a NumPy array
arr = np.array([1, 2, 3, 4, 5])
print("Array:", arr)

# Basic operations
print("Mean:", np.mean(arr))
print("Square root:", np.sqrt(arr))
print("Reshape:", arr.reshape(5,1))
```

**2. pandas (Data Handling)**

- pandas is used for **data manipulation and analysis**.  
- Provides two main structures:  
  - **Series** (1D labeled array, like a column in Excel).  
  - **DataFrame** (2D labeled data, like an Excel sheet).  
- Used in almost all data science pipelines.

**Example:**  


```python
import pandas as pd

# Create a DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "Department": ["HR", "IT", "Finance"]
}

df = pd.DataFrame(data)
print(df)

# Access column
print("Ages:", df["Age"])

# Describe statistics
print(df.describe())
```

**Task 2:**  
Create a pandas DataFrame of students with columns: Name, Marks in 3 subjects.  
Calculate the average marks and find the student with the highest score.


```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Subject1': [85, 90, 78],
    'Subject2': [88, 85, 82],
    'Subject3': [92, 87, 80]
})

df['Average'] = df[['Subject1', 'Subject2', 'Subject3']].mean(axis=1)

top_student = df.loc[df['Average'].idxmax()]

print(df)
print("\nTop student:\n", top_student)


```

**3. Matplotlib (Visualization)**

- Matplotlib is used for **data visualization**.  
- It helps create line charts, bar charts, scatter plots, histograms, etc.  
- It is the foundation of Python plotting libraries.  

**Example:**  


```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y, label="Line Plot", color="blue")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Chart")
plt.legend()
plt.show()
```

**Task 3:**  
Create a bar chart showing marks of 5 students using Matplotlib.  


```python
#Create a bar chart showing marks of 5 students using Matplotlib.
import matplotlib.pyplot as plt

x = ["Aash","Sheik","Ani","Aji","Raahman","Meeran","Ashath"]
y = [100,99,50,55,25,20,15]

plt.bar(x, y, label="Bar Chart", color="pink")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.title("Bar Chart")
plt.legend()
plt.show()
```

**4. Seaborn (Statistical Visualization)**

- Seaborn is built on top of Matplotlib.  
- Provides **high-level interface** for drawing attractive and informative statistical graphics.  
- Great for analyzing patterns and distributions.  

**Example:**  


```python
import seaborn as sns
import numpy as np

# Generate random data
data = np.random.normal(size=100)

# Create histogram with density curve
sns.histplot(data, kde=True)
plt.title("Histogram with Seaborn")
plt.show()
```

**Task 4:**  
Use Seaborn to create a scatterplot showing the relationship between "Age" and "Marks" of students.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Name': ['Aash', 'Sheik', 'Ani', 'Aji', 'Raahman', 'Meeran', 'Ashath'],
    'Age': [16, 17, 16, 15, 18, 17, 16],
    'Marks': [100, 99, 50, 55, 25, 20, 15]
}

df = pd.DataFrame(data)

sns.scatterplot(x='Age', y='Marks', data=df)

plt.title("Scatterplot of Age vs Marks")
plt.xlabel("Age")
plt.ylabel("Marks")

plt.show()
```
